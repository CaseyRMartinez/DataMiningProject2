```python
path = "D:/PatternEnergy/"
df = pd.read_csv(path + "SCADAcombined2017Q1.csv")

# df['TimeStamp'] = pd.to_datetime(df['TimeStamp'])
# df.set_index('TimeStamp', inplace=True)

# df.dropna(inplace=True)

print("Dimensions of data frame: ", df.shape)
print(df.info())
print(df.index)
df.head()
```

```python
cols = ['Name','wtc_ActPower_mean', 'wtc_AcWindSp_mean']

df = df[cols].groupby('Name',as_index=True).median() #

df.describe()
```

```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
from sklearn.preprocessing import StandardScaler

X = StandardScaler().fit_transform(df.select_dtypes(include='float64'))

plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
# generate the linkage matrix
Z = linkage(X, 'ward')

from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist

c, coph_dists = cophenet(Z, pdist(X))
print("Cophenetic Distance = ", c)

from scipy.cluster.hierarchy import fcluster

clusters = fcluster(Z, t=1, criterion='distance')

df['cluster'] = clusters
print("Number of Clusters = ",len(set(clusters)))
```

```python
for i in range(1,4):
    clusters = fcluster(Z, t=i, criterion='distance')
    print("Number of Clusters = ",len(set(clusters)))
```

```python
# Graph Clusters

y = df['wtc_ActPower_mean']
x = df['wtc_AcWindSp_mean']
c = df['cluster']

plt.figure(figsize=(15, 7))
plt.scatter(y=y, x=x, c=c, cmap='prism')  # plot points with cluster dependent colors

plt.title("Turbine Performance Clusters")
plt.xlabel('Avg. Wind Speed')
plt.ylabel('Avg. Power')
# plt.legend(loc='best')
plt.show()
```

```python
# calculate full dendrogram
plt.figure(figsize=(25, 30))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Distance')
plt.ylabel('Turbine Index')
dendrogram(
    Z,
    orientation='right',
#     leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=20.,  # font size for the x axis labels
)
plt.show()
```

```python
plt.title('Hierarchical Clustering Dendrogram (truncated)')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
    Z,
    truncate_mode='lastp',  # show only the last p merged clusters
    p=12,  # show only the last p merged clusters
    show_leaf_counts=False,  # otherwise numbers in brackets are counts
    leaf_rotation=90.,
    leaf_font_size=12.,
    show_contracted=True,  # to get a distribution impression in truncated branches
)
plt.show()
```
