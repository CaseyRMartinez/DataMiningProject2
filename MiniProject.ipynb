# Making class variables
First thought was to use OneHotEncoder. However, I wanted to first attempt a simpler approach. I identified the top 10 alarms and then sub-selected the dataset to only include those alarms. Since there were 10 variables I will attempt to use a SVM to classify the alarms.

```python
top = df_merged_alarms['Description'].value_counts()
```

```python
top = [x for x in dict(top[0:10]).keys()]
top
```

```python
df_class = df_merged[df_merged["Description"].isin(top)]
df_class
```

# Start SVM

```python
from sklearn import svm
from sklearn.preprocessing import Imputer
# missing_values is the value of your placeholder, strategy is if you'd like mean, median or mode, and axis=0 means it calculates the imputation based on the other feature values for that sample
# borrowed from https://stackoverflow.com/questions/11441751/how-to-get-svms-to-play-nicely-with-missing-data-in-scikit-learn

#Set up data
y = df_class["Code"]
ref = df_class[["Code", "Description"]].groupby("Code")
ref.describe()

#get rid of all the string variables
X = df_class.drop(["Code", "Description", "TimeStamp", "wtc_CurTime_endvalue", "DateTimeOn", 
                   "TimeOn", "TimeOff", "Source", "Name", "Fault Type"], axis=1)

#Impute NaN Data so that sklearn will play nice
imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
imp.fit(X)
train_X = imp.transform(X)


# Make test and training data
# http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
svm_class = svm.SVC(C=0.5, kernel='rbf', degree=3, gamma='auto')
svm_class.fit(X_train,y_train)

#borrowed from Machine Learning notebook 4
y_hat = svm_class.predict(X_test)
acc = accuracy_score(y_test,y_hat)
conf = confusion_matrix(y_test,y_hat)
print('accuracy:', acc )
print(conf)
```
